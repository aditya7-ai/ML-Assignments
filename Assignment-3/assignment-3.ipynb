```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn 
```

```python
ds=pd.read_csv("admission.csv")
```

```python
# Scatter Plot
plt.figure(figsize=(6,4))
plt.scatter(ds['GRE Score'], ds['CGPA'], c='green', alpha=1,)
plt.xlabel('GRE Score')
plt.ylabel('CGPA')
plt.title('Scatter Plot')
plt.show()
```

```python
# Bar Plot
plt.figure(figsize=(6,4))
plt.bar(ds['University Rating'], ds['CGPA'], color= 'orange')
plt.xlabel('University Rating')
plt.ylabel('CGPA')
plt.title('Bar Plot')
plt.show()
```

```python
# Box Plot
plt.figure(figsize=(6,4))
sn.boxplot(x=ds['University Rating'], y=ds['TOEFL Score'])
plt.title('Box Plot')
plt.show()
```

```python
sn.boxplot(x = "GRE Score", data = ds, width = 0.8, showmeans = True)
```

```python
# Pie Chart
plt.figure(figsize=(6,6))
research = ds['Research'].value_counts()
plt.pie(research, labels=['Research (1)', 'No Research (0)'], autopct='%1.1f%%', colors=['orange', 'green',])
plt.title('Pie Chart')
plt.show()
```

```python
# line chart
plt.plot(ds["Serial No."], ds["CGPA"])
plt.xlabel('Serial No.')
plt.ylabel('CGPA')
plt.title('Line Chart')
plt.show()
```
